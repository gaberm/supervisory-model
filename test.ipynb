```python
lon, lat = -86.85852778710755, 36.149286642168455
point = Point(lon, lat)
```

```python
point_gdf = gpd.GeoDataFrame(geometry=[point], crs="EPSG:4326")
point_proj = point_gdf.to_crs(links.crs).geometry.iloc[0]

# Calculate distance from point to all geometries
distances = links.geometry.distance(point_proj)

# Get the index of the nearest link
nearest_idx = distances.idxmin()

nearest_link = links.loc[nearest_idx]
nearest_link
```
